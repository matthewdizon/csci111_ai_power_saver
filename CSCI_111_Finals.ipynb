```python
import random 

"""
This part of the code is the backbone of the entire project. This is what
happens behind the scenes, depending on what the user selects. This is also
where the different variables and functions are initialized.
"""

# Initializes the tree
# This represents all the different possible states and the possible moves to make from each state
graph = {
    # Left, Right, On, Off
    '1' : ['1', '8', '1', '2'],
    '2' : ['2', '3', '1', '2'],
    '3' : ['2', '3', '3', '4'],
    '4' : ['5', '4', '3', '4'],
    '5' : ['5', '4', '6', '5'],
    '6' : ['6', '7', '6', '5'],
    '7' : ['6', '7', '8', '7'],
    '8' : ['1', '8', '8', '7']
}

def displayCurrentState(node):
  if node == "1":
    current_room = "Left"
    room_of_person = "Left"
    light_left_room = "On"
    light_right_room = "On"
  elif node == "2":
    current_room = "Left"
    room_of_person = "Left"
    light_left_room = "Off"
    light_right_room = "On"
  elif node == "3":
    current_room = "Right"
    room_of_person = "Left"
    light_left_room = "Off"
    light_right_room = "On"
  elif node == "4":
    current_room = "Right"
    room_of_person = "Left"
    light_left_room = "Off"
    light_right_room = "Off"
  elif node == "5":
    current_room = "Left"
    room_of_person = "Left"
    light_left_room = "Off"
    light_right_room = "Off"
  elif node == "6":
    current_room = "Left"
    room_of_person = "Left"
    light_left_room = "On"
    light_right_room = "Off"
  elif node == "7":
    current_room = "Right"
    room_of_person = "Left"
    light_left_room = "On"
    light_right_room = "Off"
  elif node == "8":
    current_room = "Right"
    room_of_person = "Left"
    light_left_room = "On"
    light_right_room = "Off"
    
  print(f"""
  ----- CURRENT STATE -----
  State No.: {node}
  Current Room of Butler: {current_room}
  Room of Person: {room_of_person}
  Light of Left Room: {light_left_room}
  Light of Right Room: {light_right_room}
  ----- CURRENT STATE -----
  """)

# Start of DFS Function
def dfs(node, path):
    path.append(node)
    print(f"\nStep No. {len(path)}")

    displayCurrentState(node)
    print(f"\nDequeueing State {node} for goal check")
    print("Goal Checking The Node", node)

    if len(path) == 499:
      print("We've ran through the algorithm 499 times!")
      print(f"The path of the agent so far is: {path}")
      return True

    if node in goal_state:
      print(node, goal_state)
      print("Goal Check = Success")
      return f"""
\nPower is now being saved! Goal state achieved from State {path[0]} to State {path[-1]} in {len(path)} {'steps' if len(path) > 1 else 'step'}!
This was the entire path of the agent: {path}
"""

    print("Goal Check = Fail")

    for child in graph[node]:
      if dfs(child, path):
        return "This is an endless loop... we suggest using a different algorithm such as DLS"

# End of DFS Function
# Start of BFS Function

def bfs(node, path):
  queue.append(node)

  while queue:
    node = queue.pop(0)
    path.append(node)
    print(f"\nStep No. {len(path)}")
    displayCurrentState(node)
    
    print(f"\nDequeueing State {node} for goal check")
    
    print("Goal Checking The Node", node)
    if node in goal_state:
      print("Goal Check = Success")
      return f"""
\nPower is now being saved! Goal state achieved from State {path[0]} to State {path[-1]} in {len(path)} {'steps' if len(path) > 1 else 'step'}!
This was the entire path of the agent: {path}
"""
    
    print("Goal Check = Fail")

    curr_queue = []

    for child in graph[node]:
      queue.append(child)
      curr_queue.append(child)
    print(f"Enqueuing new nodes: {curr_queue}")
    print(f"Complete Queue: {queue}")

# End of BFS Function
# Start of DLS Function

def dls(node, path, level, maxD):
  path.append(node)
  print(f"\nStep No. {len(path)}")
  displayCurrentState(node)

  print('\nCurrent Level: ', level)
  print('Goal Checking The Node', node)

  if node in goal_state:
    print("Goal Check = Success")
    return f"""
\nPower is now being saved! Goal state achieved from State {path[0]} to State {path[-1]} in {len(path)} {'steps' if len(path) > 1 else 'step'}!
This was the entire path of the agent: {path}
"""

  print('Goal Check = Fail')
  
  if level==maxD:
    return False

  print('\nExpanding the current node', node)

  for child in graph[node]:
    if dls(child, path, level+1, maxD):
      return f"""
\nPower is now being saved! Goal state achieved from State {path[0]} to State {path[-1]} in {len(path)} {'steps' if len(path) > 1 else 'step'}!
This was the entire path of the agent: {path}
"""
  return False

# End of DLS Function

"""
This part of the code focuses on user interaction. It makes sure that the user
can select set an initial states of the agent and environment. It also makes
sure that the values of the input from the user are valid.
"""

## Information about the agent and its environment
while True:
  queue = [] # Initialize a queue
  path = [] # Initializes a path that captures the entire steps taken
  initial_level = 0 # Initializes the level (for DFS/DLS)

  print("""
  ----- INTRODUCTION -----
  Hello and Welcome to Power Saver!

  In this game, the agent is a butler who aims to turn the lights on/off
  inside a household depending on the details of its environment

  The environment has 3 main details:
  1.) STATE OF DAY - The butler will check if it is morning/evening
  2.) STATE OF LIGHT - The butler will check if the light in the current room is turned on/off
  3.) STATE OF PERSON - The butler will check a person is inside the current room

  These are the conditions of the butler:
  1.) It will turn the lights off in both rooms if it is morning, regardless of having a person inside
  2.) It will turn the lights on of the room with a person inside, only if it is evening
  3.) All rooms without a person inside should have its light turned off
  ----- INTRODUCTION -----
  """)

## User Select: Random vs Manual
  print("Let's begin! Would you like to manually select the details of the initial state, or would you want to randomize it?")
  while True:
    selected_user_control = input("\nType R for Random or Type M for Manual: ")
    if selected_user_control.upper() not in ["R", "M"]:
        print("Please choose between R (Random) or M (Manual)")
    else:
      print(f"\nGreat! The selected option is: {'Random' if selected_user_control.upper() == 'R' else 'Manual'}")
      break

  if selected_user_control.upper() == "R": # User randomizes the intial states of the game
    random_day = ["M", "E"]
    state_of_day = random.choice(random_day)
    
    random_person = ["L", "R"]
    state_of_person = random.choice(random_person)

    random_state = ["1", "2", "3", "4", "5", "6", "7", "8"]
    initial_state = random.choice(random_state)

  elif selected_user_control.upper() == "M": # User manually sets the intial states of the game
    while True:
      print("\nWhat will be the STATE OF DAY? Is it MORNING or EVENING?")
      state_of_day = input("Type M for Morning or Type E for Evening: ")
      if state_of_day.upper() not in ["M", "E"]:
        print("Please choose between M (Morning) or E (Evening)")
      else:
        print(f"Great! The selected state of day is: {'Morning' if state_of_day.upper() == 'M' else 'Evening'}")
        break

    while True:
      print("\nWhat will be the STATE OF PERSON? Will the person be inside the LEFT room or the RIGHT room?")
      state_of_person = input("Type L for Left Room or Type R for Right Room: ")
      if state_of_person.upper() not in ["L", "R"]:
        print("Please choose between L (Left Room) or R (Right Room)")
      else:
        print(f"Great! The selected state of person is: {'Left Room' if state_of_person.upper() == 'L' else 'Right Room'}")
        break

    while True:
      print("\nWhich state will the agent start in? This determines which room the butler starts in and if the lights are turned on/off")
      initial_state = input("Select a number from (1-8) to select the Initial State: ")
      if initial_state not in ["1", "2", "3", "4", "5", "6", "7", "8"]:
        print("Please choose a number from (1-8)")
      else:
        print(f"Great! The selected initial state is: {initial_state}")
        break

  # Prints the details of the agent's environment
  print(f"""
  \n
  ----- ENVIRONMENT DETATILS ----- 
  State of Day: {"Morning" if state_of_day.upper() == "M" else "Evening"}
  Room The Person is Inside: {"Left Room" if state_of_person.upper() == "L" else "Right Room"}
  Initial Room of The Butler: {initial_state}
  ----- ENVIRONMENT DETATILS -----""")

  ## Sets goal state of the agent
  if state_of_day.upper() == "M":
    goal_state = ["4", "5"]
  elif state_of_day.upper() == "E":
    if state_of_person.upper() == "L":
      goal_state = ["6", "7"]
    elif state_of_person.upper() == "R":
      goal_state = ["2", "3"]

  print(f"\nThese are the possible goal states of the agent: {goal_state}")

  ## User Select: DFS vs BFS
  print("\nAwesome work! Up next, you'll have to select an algorithm for the agent to use. ")
  while True:
    selected_algorithm = input("\nType D1 for DFS or Type B for BFS or Type D2 for DLS: ")
    if selected_algorithm.upper() not in ["D1", "B", "D2"]:
        print("Please choose between D1 (DFS) or B (BFS) or D2 (DLS)")
    else:
      print(f"Great! The selected algorthim is: {'DFS' if selected_algorithm.upper() == 'D1' else 'DLS' if selected_algorithm.upper() == 'D2' else 'BFS'}")
      break
  if selected_algorithm.upper() == "D1":
    run_algorithm = dfs(initial_state, path)
  elif selected_algorithm.upper() == "B":
    run_algorithm = bfs(initial_state, path)
  elif selected_algorithm.upper() == "D2":
    # User selects max depth for DLS algorithm
    while True:
        max_depth = input("\nPlease enter a number (1-900) to specify the MAX DEPTH for the algorithm: ")
        try:
            if int(max_depth) in list(range(1,901)):
              val = int(max_depth)
              print("Max Depth is: ", val)
              break
            else:
              print("Please choose a number from (1-900)")
        except:
            print("Please choose a number from (1-900)")

    run_algorithm = dls(initial_state, path, initial_level, val)

  print(run_algorithm)

  while True: # Identifies if the user wants to re-run the program
      answer = input('\nRun again? (Y/N): ')
      if answer.upper() in ('Y', 'N'):
          break
      print("Please choose between Y (Yes) or N (No).")
  if answer.upper() == 'Y':
      continue
  else:
      print("Program has been stopped. Goodbye!")
      break
```
